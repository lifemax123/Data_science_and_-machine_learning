### Plot line 

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([0,1,2,3,4,5,6,7,9,10])
y = np.array([1,2,5,10,17,26,37,50,82,101])

plt.plot(x,y)
plt.grid()
fig = plt.figure()
plt.show()

print('60 to 70')
```

### Matplotlib increase 

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0,40,2)
y = 2**x

plt.plot(x,y,'--',color ='blue')
plt.grid()
plt.show()
print(30)
```

### Plot Graphs

```python
import matplotlib.pyplot as plt

import numpy as np

x = np.arange(0,40,2)
y = x

x1 = [0,1,2,3,4,5,6,7,9,10]
y1 = [1,2,5,10,17,26,37,50,82,101]

plt.plot(x1, y1, label='x[i]=x[i]^2 +1')
plt.plot(x, y, label='x[i]=y[i]')
plt.ylabel('Y')
plt.xlabel('X')
plt.legend()
plt.show()
```

### Microsoft Bubble Graph

### Laptop Sales

```python
import numpy as np
import matplotlib.pyplot as plt
company = ['HP','Dell','Lenovo','Asus','Apple','Acer']
sold = [20000,43000,15000,17000,22000,13000]

company = np.array(company)
sold = np.array(sold)

plt.pie(sold, labels=company, autopct='%1.1f%%', counterclock=False)
plt.axis('equal')
plt.show()

x = np.true_divide(sold, sold.sum(axis=0, keepdims=True))*100
for i in range(len(company)):
    print(company[i],format(x[i],'.1f'))
```
